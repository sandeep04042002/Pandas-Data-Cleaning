```python
def categorize_age(age):
    if age < 13:
        return 'Child'
    elif 13 <= age < 20:
        return 'Teen'
    elif 20 <= age < 65:
        return 'Adult'
    else:
        return 'Senior'
df['Age_Group'] = df['Age'].apply(categorize_age)
display(df[['Age', 'Age_Group']].head())
```

```python
def categorize_loyalty(total_purchase):
    if total_purchase > 10000:
        return 'Platinum'
    elif total_purchase > 5000:
        return 'Gold'
    elif total_purchase > 1000:
        return 'Silver'
    else:
        return 'Bronze'

df['Loyalty_Level'] = df['Total_Purchase_Amount'].apply(categorize_loyalty)
display(df[['Total_Purchase_Amount', 'Loyalty_Level']].head())
```
